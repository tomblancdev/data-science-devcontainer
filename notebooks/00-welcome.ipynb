```python
import duckdb
import polars as pl

df = pl.DataFrame(
    {
        "city": ["Paris", "Lyon", "Paris", "Lyon", "Nice"],
        "sales": [120, 90, 80, 30, 60],
    }
)
df
```

```python
# Polars: lazy, expression-based group-by
df.group_by("city").agg(pl.col("sales").sum().alias("total")).sort("total", descending=True)
```

```python
# DuckDB: query the Polars DataFrame directly with SQL (zero-copy via Arrow)
duckdb.sql(
    """
    select city, sum(sales) as total
    from df
    group by city
    order by total desc
    """
).pl()
```
